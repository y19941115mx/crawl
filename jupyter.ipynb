```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
c = MongoClient()
cursor = c.weixin['post'].find()
df = pd.DataFrame(list(cursor))
df.drop('_id', axis=1)
display_columns = ['title','read_num','like_num', 'p_date']
df.reindex(columns=display_columns)
```

```python
df.describe()
```

```python
top_read_num = df.sort_values(by=['read_num'], ascending=False)[:10][display_columns]
top_like_num = df.sort_values(by=['like_num'], ascending=False)[:10][display_columns]
```

```python
top_read_num
```

```python
top_like_num
```

```python
ax = df.plot(y='read_num', x='p_date', title='文章阅读趋势')
ax.set_ylabel('read_num')
ax.set_xlabel('')
```

```python
ax = top_like_num.plot(x='title', y='like_num',title='文章点赞数TOP10', kind='barh', figsize=(9,6),fontsize=15)
ax.set_ylabel('')
ax.set_xlabel('赞赏数')
ax.legend().set_visible(False)
```

```python
ax = top_read_num.plot(x='title', y='read_num',title='文章阅读量TOP10', kind='barh', figsize=(9,6),fontsize=15)
ax.set_ylabel('')
ax.set_xlabel('阅读数')
ax.legend().set_visible(False)
```

### 按年分组统计每年写文章的数量

```python
year_df = df.groupby(df.p_date.dt.year).size().reset_index(name='total')
year_df
```

```python
ax = year_df.plot(x='p_date', y='total',title='写文章的数量', kind='bar', figsize=(9,6),fontsize=15)
ax.set_ylabel('文章数')
ax.set_xlabel('')
ax.legend().set_visible(False)
```
